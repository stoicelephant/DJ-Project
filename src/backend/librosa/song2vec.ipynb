```python
import librosa 
import numpy as np 
import matplotlib.pyplot as plt 
import librosa.display
```

```python
filename = 'test_file.mp3'
y,sr = librosa.load(filename)
song_length = librosa.get_duration(y=y,sr=sr)
# the song length in seconds is 217
#now song length x sr should be equal to the length of y, because 
#for every second we are sampling it sr-times
#print(len(y) == song_length*sr) #returns True 

segment = y[:librosa.time_to_samples(5)] #looking at a 5 second window 
# we could also do this using utility functions 

S =  np.abs(librosa.stft(segment, n_fft = 512, hop_length=256))

S.shape
```
